```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn import preprocessing
```

```python
iris = datasets.load_iris()
x = pd.DataFrame (iris.data)
x.columns = ['sepal_length', 'sepal_width','petal_length', 'petal_width']
y = pd.DataFrame(iris.target)
y.columns = ['target']
color_map = np.array (['red', 'lime', 'black'])
```

## K-Means

```python
knn_model = KMeans (n_clusters = 3)
knn_model.fit (x)
```

# EM Gaussian Mixtrue algorithm

```python
scaler = preprocessing.StandardScaler().fit (x)

xsa = scaler.transform (x)

xs = pd.DataFrame (xsa, columns = x.columns)

gmm = GaussianMixture (n_components = 3)

gmm.fit (xs)

gmm_y = gmm.predict (x)
```

```python
#plotting real clusters
plt.subplot (1, 2, 1)
plt.scatter (x.petal_length, x.petal_width, c = color_map[y.target], s = 40)
plt.title ("petal real scatter")
plt.xlabel ("petal length")
plt.ylabel ("petal width")

#plotting real clusters
plt.subplot (1, 2, 2)
plt.scatter (x.sepal_length, x.sepal_width, c = color_map[y.target], s = 40)
plt.title ("sepal real scatter")
plt.xlabel ("sepal length")
plt.ylabel ("sepal width")
```

```python
#k means plotting
#plotting KNN clusters
plt.subplot (1, 2, 1)
plt.scatter (x.petal_length, x.petal_width, c = color_map[knn_model.labels_], s = 40)
plt.title ("petal knn clustering")
plt.xlabel ("petal length")
plt.ylabel ("petal width")

#plotting real clusters
plt.subplot (1, 2, 2)
plt.scatter (x.petal_length, x.petal_width, c = color_map[y.target], s = 40)
plt.title ("petal real scatter")
plt.xlabel ("petal length")
plt.ylabel ("petal width")
```

```python
#k means plotting
#plotting gmm clusters
plt.subplot (1, 2, 1)
plt.scatter (x.petal_length, x.petal_width, c = color_map[gmm_y], s = 40)
plt.title ("petal gmm clustering")
plt.xlabel ("petal length")
plt.ylabel ("petal width")

#plotting real clusters
plt.subplot (1, 2, 2)
plt.scatter (x.petal_length, x.petal_width, c = color_map[y.target], s = 40)
plt.title ("petal real scatter")
plt.xlabel ("petal length")
plt.ylabel ("petal width")
```

# Practice 2

```python
import matplotlib.pyplot as plt
import sklearn
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn import preprocessing
import sklearn.metrics as sm
import numpy as np
import pandas as pd
```

```python
iris = datasets.load_iris()
x = pd.DataFrame (iris.data)
y = pd.DataFrame (iris.target)
x.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
y.columns = ['targets']
color_map = np.array (['red', 'lime', 'black'])
```

```python
plt.subplot (1, 2, 1)
plt.scatter (x.petal_length, x.petal_width, c = color_map[y.targets])
plt.title ("Petal clusturing - real")

plt.subplot (1, 2, 2)
plt.scatter (x.sepal_length, x.sepal_width, c = color_map[y.targets])
plt.title ("Sepal clusturing - real")
```

```python
#knn
model1 = KMeans (n_clusters = 3)
model1.fit (x)

print ("accuracy knn = ", sm.accuracy_score(y, model1.labels_))

plt.subplot (1, 2, 1)
plt.scatter (x.petal_length, x.petal_width, c = color_map[model1.labels_])
plt.title ("Petal clusturing - knn")

plt.subplot (1, 2, 2)
plt.scatter (x.sepal_length, x.sepal_width, c = color_map[model1.labels_])
plt.title ("Sepal clusturing - knn")
```

```python
#gmm
scaler = preprocessing.StandardScaler().fit(x)
xsa = scaler.transform(x)
xs = pd.DataFrame (xsa, columns = x.columns)
gmm = GaussianMixture(n_components = 3)
gmm.fit (xs)
gmm_y = gmm.predict (xs)

print ("accuracy knn = ", sm.accuracy_score(y, gmm_y))

plt.subplot (1, 2, 1)
plt.scatter (x.petal_length, x.petal_width, c = color_map[gmm_y])
plt.title ("Petal clusturing - gmm")

plt.subplot (1, 2, 2)
plt.scatter (x.sepal_length, x.sepal_width, c = color_map[gmm_y])
plt.title ("Sepal clusturing - gmm")
```
